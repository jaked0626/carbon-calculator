````python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds
````

### Retrieve Food101 Dataset using TFDS

````python
(train_data, test_data), info = tfds.load(name='food101', 
                                          split=['train', 'validation'], 
                                          shuffle_files=True,
                                          as_supervised=True, # Retrieves data in tuple format
                                          with_info=True) # Returns dataset metadata
````

### Data Preprocessing

1. Reshapes image to [img_shape, img_shape, color_channels]
2. Converts image datatype from *uint8* -> *float32*

````python
def preprocess(image, label, img_shape=224):
    image = tf.image.resize(image, [img_shape, img_shape]) # reshape to img_shape
    return tf.cast(image, tf.float32), label # return (float32_image, label) tuple
````

````python
for image, label in train_data.take(1):
    fig = plt.figure()
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)

    print(f"Label: {info.features['label'].names[label]}")
    
    ax1.imshow(image.numpy())
    ax1.axis('off')
    ax1.set_title('Original Image')
    
    img, lbl = preprocess(image, label)
    ax2.imshow(img/255.)
    ax2.axis('off')
    ax2.set_title('Processed Image')
````

### Batch Dataset

Computing on batches are more memory efficient.
- ```map()``` maps a predefined function to a target dataset
    - ```num_parallel_calls``` parallelizes preprocessing and significantly improves the speed!
- ```shuffle()``` randomly shuffles the elements of a target dataset
- ```batch()``` converts datasets into small batches
- ```prefetch()``` prepares subsequent batches of data whilst other batches of data are being computed on

````python
# Map preprocessing function to training data (and paralellize)
train_data = train_data.map(map_func=preprocess, num_parallel_calls=tf.data.AUTOTUNE)
# Shuffle train_data and turn it into batches and prefetch it (load it faster)
train_data = train_data.shuffle(buffer_size=1000).batch(batch_size=32).prefetch(buffer_size=tf.data.AUTOTUNE)

# Map prepreprocessing function to test data
test_data = test_data.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
# Turn test data into batches (don't need to shuffle)
test_data = test_data.batch(32).prefetch(tf.data.AUTOTUNE)
````

### Model Callbacks

````python
checkpoint_path = 'model_checkpoints/cp.ckpt' # saving weights in '.ckpt' file
model_checkpoint = tf.keras.callbacks.ModelCheckpoint(checkpoint_path,
                                                      monitor='val_accuracy',
                                                      save_best_only=True,
                                                      save_weights_only=True, # Saves only the weights
                                                      verbose=1)
````

````python
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing

# Create base model
input_shape = (224, 224, 3)
base_model = tf.keras.applications.EfficientNetB0(include_top=False)
base_model.trainable = False # freeze base model layers

# Create Functional model 
inputs = layers.Input(shape=input_shape, name="input_layer", dtype=tf.float16)
# Note: EfficientNetBX models have rescaling built-in but if your model didn't you could have a layer like below
# x = preprocessing.Rescaling(1./255)(x)
x = base_model(inputs, training=False) # set base_model to inference mode only
x = layers.GlobalAveragePooling2D(name="pooling_layer")(x)
x = layers.Dense(len(info.features['label'].names))(x) # want one output neuron per class 
# Separate activation of output layer so we can output float32 activations
outputs = layers.Activation("softmax", dtype=tf.float32, name="softmax_float32")(x) 
model = tf.keras.Model(inputs, outputs)

# Compile the model
model.compile(loss="sparse_categorical_crossentropy", # Use sparse_categorical_crossentropy when labels are *not* one-hot
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["accuracy"])
````

````python
model.summary()
````

````python
# Fit the model with callbacks
with tf.device('/GPU:0'):
    food101 = model.fit(train_data, 
                        epochs=3,
                        steps_per_epoch=len(train_data),
                        validation_data=test_data,
                        validation_steps=int(0.15 * len(test_data)),
                        callbacks=[model_checkpoint])
````
