```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds
```

### Retrieve Food101 Dataset using TFDS

```python
(train_data, test_data), info = tfds.load(name='food101', 
                                          split=['train', 'validation'], 
                                          shuffle_files=True,
                                          as_supervised=True, # Retrieves data in tuple format
                                          with_info=True) # Returns dataset metadata
```

# Plot an Image from TDFS

```python
for image, label in train_data.take(1):
    plt.imshow(image.numpy())
    plt.axis('off')
    print(f"Label: {info.features['label'].names[label]}")
```
